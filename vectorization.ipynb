```python
import time

a = np.random.rand(10000000)
b = np.random.rand(10000000)
print(a)
print(b)
tic = time.time()
c=np.dot(a,b)
toc= time.time()
print(c)
print("Vectorize version: "+ str(1000*( toc -  tic))+ "ms")
c=0
tic=time.time()

for i in range(10000000):
    c+= a[i]* b[i]
toc = time.time()
print(c)
print("UnVectorize version: "+ str(1000*( toc -  tic))+ "ms")
```
